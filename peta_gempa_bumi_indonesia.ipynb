```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#data lokasi gempa bumi indonesia
gempa = pd.read_csv("lokasi_gempa_bumi_indonesia.csv")
gempa.head()
```

```python
#Tidak terjadi tsunami
TidakTsunami = gempa[gempa["Tsu"] == "Tidak Terjadi Tsunami"]
TidakTsunami.head()
```

```python
#Terjadi tsunami
TerjadiTsunami = gempa[gempa["Tsu"] == "Terjadi Tsunami"]
TerjadiTsunami.head()
```

```python
#ukuran gambar peta berdasarkan LatLong 
BBox = ((91.23, 141.59, -13.41, 11.44))
BBox
```

```python
#gambar peta diambil dari https://www.openstreetmap.org/export#map=5/-1.430/117.465 
#dengan area (91.23, 141.59, -13.41, 11.44)
peta = plt.imread("indonesia.jpg")
```

```python
#menampilkan scatterplot
fig, ax = plt.subplots(figsize = (15,15))
ax.scatter(TidakTsunami.Longitude, TidakTsunami.Latitude, zorder=1, alpha= 0.4, c='b', s=20)
ax.scatter(TerjadiTsunami.Longitude, TerjadiTsunami.Latitude, zorder=2, alpha= 0.4, c='r', s=20)
ax.set_title('Plotting Spatial Data on Indonesia Map')
ax.set_xlim(91.23,141.59)
ax.set_ylim(-13.41,11.44)
ax.imshow(peta, zorder=0, extent = BBox, aspect= 'equal')
```
